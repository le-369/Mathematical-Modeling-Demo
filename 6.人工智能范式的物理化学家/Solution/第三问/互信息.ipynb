```python
import pandas as pd
from sklearn.feature_selection import mutual_info_regression
import numpy as np

data = pd.read_excel('features1.xlsx')
```

```python
data.head()
```

```python
# 定义你感兴趣的特征列
selected_features = [
    'x95', 'x65', 'x89', 'x44', 'x4', 'x7', 'x6', 'y1', 'x99', 'x21', 'x53',
    'x86', 'x80', 'x85', 'x78', 'x63', 'x73', 'x61', 'x51', 'x36', 'x17', 
    'x97', 'x98', 'x1', 'x59', 'x69', 'x27', 'x55', 'x92', 'x15', 'x10',
    'x79', 'x57', 'x90', 'x71', 'x75', 'x3', 'x29', 'x74', 'x52'
]

# 提取特征矩阵X（除y3以外的特征）
X = data[selected_features]
# 提取目标变量y3
y = data['y3']
```

```python
# 计算互信息
mi_scores_selected = mutual_info_regression(X, y)

# 将结果与特征名配对
mi_scores_df = pd.DataFrame({
    'Feature': selected_features,
    'Mutual Information Score': mi_scores_selected
})

mi_scores_df.to_excel('mi_scores.xlsx', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 设置绘图风格
sns.set(style="whitegrid")

# 创建绘图区域
plt.figure(figsize=(10, 6))

# 绘制条形图
sns.barplot(x='Mutual Information Score', y='Feature', data=mi_scores_df, palette="viridis")

# 设置标题和标签
plt.title('Mutual Information Scores for Features (Excluding y3)', fontsize=16)
plt.xlabel('Mutual Information Score', fontsize=14)
plt.ylabel('Features', fontsize=14)

# 显示图形
plt.show()
```
